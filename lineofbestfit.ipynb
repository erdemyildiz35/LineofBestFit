```python
import matplotlib.pyplot as plt
import numpy as np

x = [1, 2, 3, 4, 5,6,7,8]
y = [50, 65, 55, 94, 95,100,90,88]

slope, intercept = np.polyfit(x, y, 1)


x_reg = np.linspace(min(x), max(x), 100)  


y_reg = slope * x_reg + intercept


plt.scatter(x, y, label='Data Points')
plt.plot(x_reg, y_reg, label='Line of best fit')


plt.xlabel('hours')
plt.ylabel('grades')
plt.title('hours-Test scores')

plt.legend()

plt.show()
```
